# Wichtige Verteilungsfunktionen


## Normalverteilung (z-Verteilung)

Die Gaußverteilung ist gegeben durch

$$p(x)=\frac{1}{\sigma_x\sqrt{2\pi}}e^{-\frac{(x-\mu_x)^2}{2\sigma_x^2}}$$

Es gilt

$$E[x]=\mu_x$$ 

und

$$E[(x-\mu_x)^2]=\sigma_x^2$$

Substitution von $z=\frac{x-\mu_x}{\sigma_x}$ liefert die standardisierte Normalverteilung

$$p(z)=\frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}$$

für die gilt $E[x]=\mu_z=0$ und  $E[(x-\mu_z)^2]=\sigma_z^2=1$

Die Wahrscheinlichkeit $P$ berechnet sich aus dem Integral

$$P(z_\alpha)=\int_{-\infty}^{z_\alpha} p(z)dz=Prob[z\lt z_\alpha]=1-\alpha$$

bzw. 

$$1-P(z_\alpha)=\int_{z_\alpha}^{\infty} p(z)dz=Prob[z \gt z_\alpha]=\alpha$$

## Normalverteilung: scipy.stats.norm(x)

```python
%pylab inline
%config InlineBackend.figure_format = 'svg'
```

## $\chi^2$-Verteilung

Gegeben sind $n$ unabhängige z-verteilte Zufallsvariablen $z_1, z_2,...,z_n$. Die Summe bildet die Zufallsvariable

$$\chi^2=z_1^2+z_2^2+...+z_n^2$$ mit $n$-Freiheitsgraden. Die Wahrscheinlichkeitsdichteverteilung für $\chi^2\gt0$ ist gegeben durch

$$p(\chi)=\frac{1}{2^{n/2}\Gamma (n/2)}\chi^{(n/2-1)}e^{-\frac{x}{2}}$$
mit der Gamma-Funktion $\Gamma (n/2)$. 

Es gilt $$E[\chi_n^2]=\mu_{\chi^2}=n$$ und  $$E[(\chi_n^2-\mu_{\chi^2})^2]=\sigma_{\chi^2}^2=2n$$

Die Gamma-Funktion ist definiert als

$$\Gamma(z) = \int_0^\infty t^{n - 1}e^{-t}\,dt$$

Für positive Integer $n$ ist das Ergebnis $\Gamma(n)=(n-1)!$.

## $\chi^2$-Verteilung: scipy.stats.chi2(x,df)

```python
x=linspace(-1,20,1000)
n=3# Freiheitsgrade
figure(figsize=(10,3))
subplot(1,2,1)
plot(x,stats.chi2.pdf(x,n))
subplot(1,2,2)
plot(x,stats.chi2.cdf(x,n))
```

## F-Verteilung

Gegeben seien die Zufallsvariablen $y_1$ und $y_2$ mit $\chi^2$-Verteilung und $n_1$ bzw. $n_2$ Freiheitsgraden. Die F-Verteilung ergibt sich aus dem Verhältnis

$$F_{n_1,n_2}=\frac{y_1/n_1}{y_2/n_2}=\frac{y_1 n_2}{y_2 n_1}$$



## Stundent $t$-Verteilung
Gegeben sind zwei unabhängige Zufallsvariablen $y$ und $z$. $y$ folgt einer $\chi^2_n$-Verteilung und $z$ ist normalverteilt. Die Student $t-$Verteilung errechnet sich aus 

$$t_n=\frac{z}{\sqrt{y/n}}$$


Es gilt $$E[t_n]=\mu_{\chi^2}=0$$ und  $$E[(t_n-\mu_t)^2]=\sigma_t^2=\frac{n}{n-2}$$ für $n>2$.



```python
x=linspace(-6,6,121)
n=3 # Freiheitsgrade
figure(figsize=(10,3))
subplot(1,2,1)
plot(x,stats.t.pdf(x,n),label='t(n='+str(n)+')')
plot(x,stats.norm.pdf(x),label='z')
legend()
subplot(1,2,2)
plot(x,stats.t.cdf(x,n),label='t(n='+str(n)+')')
plot(x,stats.norm.cdf(x),label='z')
legend(loc=4)
```

# Stichproben-Verteilungsfunktionen

## Stichproben-Mittel bei bekannter Varianz




$N$ unabhängige Messungen $x$ (normalverteilt) mit Mittelwert $\mu_x$ und Varianz $\sigma_x^2$

$$\bar{x}=\frac{1}{N} \sum_{i=1}^N x_i $$

Der Mittelwert beträgt

$$\mu_{\bar{x}}=\mu_x$$

mit Varianz

$$\sigma_{\bar{x}}^2=\frac{\sigma_x^2}{N}$$

Der Fehler (gegeben durch die Standardabweichung $\sigma$) reduziert sich proportional zu $\frac{1}{\sqrt{N}}$



## Stichproben-Varianz

Die Stichproben-Varianz ist gegeben als

$$s^2=\frac{1}{N-1} \sum_{i=1}^N (x_i-\bar{x})^2 $$

Für $N$ unabhängige Messungen $x$ (normalverteilt) mit Mittelwert $\mu_x$ und Varianz $\sigma_x^2$ gilt

$$\sum_{i=1}^N (x_i-\bar{x})^2=\sigma_x^2\chi_n^2$$
mit der $\chi_n^2$-Verteilung mit $n=N-1$ Freiheitsgraden. Es folgt für die Wahrscheinlichkeit 

$$Prob[s^2>\frac{\sigma_x^2 \chi_{n,\alpha}^2}{n} ]=\alpha$$



# Auswahl spezieller Prüfverfahren

## Vergleich zweier Mittelwerte 
Die Stichproben-Mittelwerte $\bar{a}$ und $\bar{b}$ sollen auf signifikanten Unterschied geprüft werden.  Bei gleichem Stichprobenumfang $n_a=n_b$ errechnet sich die Wahrscheinlichkeit aus

$$\hat{t}=\frac{|\bar{a}-\bar{b}|\sqrt{n}}{s_a^2+s_b^2}$$

mit der Anzahl der Freiheitsgrade beträgt $\Phi=2n-2$ ($n=n_a=n_b$)

## Vergleich Stichproben-Mittel mit bekanntem Mittelwert
Der Stichproben-Mittelwert $\bar{a}$ soll auf signifikanten Unterschied hinsichtlich des bekannten Mittelwertes $\mu$ geprüft werden.  

$$\hat{t}=|\bar{a}-\mu| \frac{\sqrt{n}}{s^2}$$

Die Anzahl der Freiheitsgrade beträgt $\Phi=n-1$.



### Beispiel
Beispiel aus Schönwiese Seite 138.

Ein geowissenschaftlicher zu untersuchender Prozess folge in guter Näherung der Normalverteilung mit $\mu$=10 und $\sigma$=2.7 Maßeinheiten. Es ist zu prüfen, ob die Stichprobe mit $\bar{a}=7.5$ und s=1.8 damit vereinbar ist, wobei der Stichproben-Umfang $n=50$ betragen soll. Der [Tabellenwert](http://de.wikipedia.org/wiki/Studentsche_t-Verteilung) für das t-Quantil beträgt $t(\Phi=49, \alpha=0.05)\approx1.68$. 

```python
mu=10.0
s=2.7

a=7.5
sa=1.8

n=50.0
df=n-1

t=abs(a-mu)*sqrt(n)/s
print t
```

```python
a=0.05
df=n-1
x=linspace(-10,0,10000)
print abs(x[argmax(stats.t.cdf(x,df)>a)])
print abs(x[argmax(stats.norm.cdf(x)>a)]) # Für große n ist das t-Quantil ähnlich der Normalverteilung
```

### Folgerung

Es ist 6.54> 1.67 

Daraus folgt, dass der Messwert nicht mit dem Prozess vereinbar ist.

# Quantile

```python
# Quantile (Verteilunfsfunktion) der Student-Verteilung (tV)
# Vergleiche Schönwiese A.3 (S. 287)
# oder http://de.wikipedia.org/wiki/Studentsche_t-Verteilung

a=0.05
x=linspace(-10,0,10000)
for df in range(1,10): #Freiheitsgrade
    print df," %1.3f " % abs(x[argmax(stats.t.cdf(x,df)>a)])
print '...'   
df=49
print df," %1.3f " % abs(x[argmax(stats.t.cdf(x,df)>a)])
print '----------------------------------------------------------------------------'
print 'Vergleich von t-Verteilung und Normalverteilung für große Anzahl Freiheitsgrade'
df=10000
print df," %1.3f " % abs(x[argmax(stats.t.cdf(x,df)>a)])

print df," %1.3f " % abs(x[argmax(stats.norm.cdf(x)>a)])
```


## Prüfung auf Daten-Unabhängigkeit

Vorsicht ist geboten, wenn die Daten nicht unabhängig sind. Dies ist der Fall, wenn z.B. Trends oder Autokorrelationen vorliegen. Weisen die Daten Autokorrelationen auf, sind sie nicht mehr unabhängig! Die Anzahl der Freiheitsgrade reduziert sich. Dies muss bei der Berechnung der Wahrscheinlichkeit berücksichtigt werden.
