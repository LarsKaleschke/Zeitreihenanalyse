## Einlesen von NetCDF-Dateien

[Network Common Data Format (NetCDF) ist ein Dateiformat für den Austausch wissenschaftlicher Daten.](http://de.wikipedia.org/wiki/NetCDF)

## Beispiel Meeresspiegel-Daten

Datenquelle ICDC: http://icdc.zmaw.de/daten/ocean/ar5-slr.html

Wie komme ich an die Daten für das Beispiel unten? Hier eine Anleitung:


 * NEU: Einzelne Komponenten ansehen und summieren: time series

 * Häkchen bei "Line plots: time series"

 * "Save as"
 
 * Select Format NetCDF

```python
%pylab inline
import scipy.io as io # Importieren eines Moduls zum Einlesen von Netcdf
```

```python
fid=io.netcdf_file('Sea_Level_AR5_RCP85.nc')
fid.variables # Welche Variablen sind in der Datei enthalten?
```

```python
SL_object=fid.variables['SELECTED_COMPONENTS'] 
# Diese Variable enthält in diesem Fall die tatsächlichen Daten
```

```python
SL_object.units # Welche Einheit?
```

```python
SL_object.data.shape # Welche Dimension?
```

```python
SL=SL_object.data[:,0,0] # Reduktion der Dimension auf eine Komponente
# Dimensionen 2 und 3 sind nicht notwendig und stören
```

```python
figure(figsize=(10,16))
plot(linspace(2007,2100,94),SL)
ylabel('Sea level [m]')
title('Global Sea Level Anomaly IPCC AR5 RCP8.5')
```
