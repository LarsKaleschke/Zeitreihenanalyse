```python
%pylab inline
%config InlineBackend.figure_format = 'svg'

from mpl_toolkits.basemap import Basemap, addcyclic
import pandas as pd

def read_buoys():
    filename='CliSAP-Buoys-20140522.csv.gz'
    #00 Asset Name, #01 Modem Id, #02 Date(GMT)
    #03 GPS LATITUDE (DEGREES), #04 GPS LONGITUDE (DEGREES),#05 FORMAT ID ,
    #06 YEAR,#07 MONTH,#08 DAY (day),#09 HOUR (hr),#10 MINUTE (min),
    #11 Sea Surface Temperature (C),
    #12 Barometric Pressure (mbar),# nur BUOY_03, BUOY_14, BUOY_16  Druck + Temperatur OK
    #13 Barometric Pressure Tendency (mbar),
    #14 Battery Voltage (V),#15 Time Since Last GPS fix (min),#16 GPS reported Signal to Noise ratio (dB),
    #17 Time to First fix (s),#18 Iridium Transmission Duration (s),#19 Hex Data
    D=pd.read_csv('CliSAP-Buoys-20140522.csv.gz',encoding = "ISO-8859-1")
    buoys=list(unique(D['Asset Name']))
    buoys.remove('Asset Name')
    DD={}
    for i,b in enumerate(buoys):
        index_buoy=D['Asset Name']==b
        Db=D[index_buoy]
        N=Db.shape[0]
        dates=dates=pd.to_datetime(Db['Date(GMT)'])
        lat=zeros(N,dtype=float)
        lon=zeros(N,dtype=float)
        # Convert degrees to float
        for j,l in enumerate(Db['GPS LATITUDE (DEGREES)'][::-1]):
            l=l.replace("'","")
            lat_deg_min=l.split('\xb0 ')
            lat[j]=float(lat_deg_min[0])+float(lat_deg_min[1])/60.0
        for j,l in enumerate(Db['GPS LONGITUDE (DEGREES)'][::-1]):
            l=l.replace("'","")
            lon_deg_min=l.split('\xb0 ')
            lon[j]=float(lon_deg_min[0])+float(lon_deg_min[1])/60.0
        lat_ts=pd.Series(lat,index=dates)
        lon_ts=pd.Series(lon,index=dates)
        lat_ts=lat_ts.resample('15min',fill_method='ffill')
        lon_ts=lon_ts.resample('15min',fill_method='ffill')
        DD[b]={'lat':lat_ts,'lon':lon_ts}        
    return DD


def plot_traj(B,b,m,s):
    lat= B[b]['lat']
    lon= B[b]['lon']
    x,y=m(lon.values,lat.values)
    m.plot(x,y,s,lw=2,label=b)
    return

#######################################################
B=read_buoys()

figure(figsize=(7,8))
m = Basemap(width=600000,height=700000,resolution='h',projection='laea',lat_ts=80,lat_0=77,lon_0=25.0)
bs=['02','03','06','09','13','14','16']# Nr. Bojen
c='rycwbgk' # Color
for i,b in enumerate(bs):
    plot_traj(B,'BUOY_'+b,m,c[i]+'-')

m.drawmapboundary(fill_color='aqua')
m.drawparallels(range(-90,100,1), color='#505050')
m.drawmeridians(range(-180,180,5), color='#505050')
m.drawcoastlines(linewidth=1.0, color='#000000')
m.fillcontinents()
legend(fontsize=6)
savefig('Buoys_trajectories.png')
```

```python
s,e='2014-03-15','2014-05-15'
b='BUOY_16'
d = {'Lon': B[b]['lon'][s:e], 'Lat': B[b]['lat'][s:e]}
index=B[b]['lon'][s:e].index
TABELLE=pd.DataFrame(data=d,index=index)
TABELLE.to_csv('CliSAP_Boje_16.csv')
```
