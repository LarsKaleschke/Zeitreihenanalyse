# Normalverteilung

```python
%pylab inline
def Normalverteilung(x,mu,sigma):
    return 1.0/(sigma*sqrt(2*pi))*exp(-(1.0/2)*(x-mu)**2/sigma**2)

N=10000
x=randn(N)
figure(figsize=(10,6))
hist(x,bins=30,range=[-5,5])

xn=linspace(-5,5)
sigma=std(x)
mu=mean(x)
y=Normalverteilung(xn,mu,sigma)*N/(30/10)
plot(xn,y,'r-',linewidth=3)
```
