#Normalverteilung


```python
%pylab inline

from __future__ import division 
# Vermeidet Integer Division, nicht notwendig für Python3

def Normalverteilung(x,mu,sigma):
    return 1.0/(sigma*sqrt(2*pi))*exp(-(1.0/2)*(x-mu)**2/sigma**2)

N=10000
x=randn(N)
figure(figsize=(10,6))
hist(x,bins=30,range=[-5,5])

xn=linspace(-5,5)
sigma=std(x)
mu=mean(x)
y=Normalverteilung(xn,mu,sigma)*N/(30/10)
plot(xn,y,'r-',linewidth=3)
```

#Übung 1
Wie viele Datenpunkte (in Prozent) liegen im Bereich $\pm1\sigma$ ?

```python
print(100-(sum(x>sigma)+sum(x<-sigma))/N*100)
```
